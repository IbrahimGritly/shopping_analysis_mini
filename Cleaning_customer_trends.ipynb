```python
# Fill missing review ratings with median per category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median())) 
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_') # Standardize column names
# replace spaces with underscores and convert to lowercase
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})  # Rename specific column
```

# Create a new column `age_group`

```python
# create an `age_group` column from `age`
#bins = [0, 17, 25, 35, 45, 55, 200]
#labels = ['0-17', '18-25', '26-35', '36-45', '46-55', '56+']
#df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, include_lowest=True).astype('category')

# quick check
#df['age_group'].value_counts().sort_index()

labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior'] # Define labels for age groups
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels) # Create age groups using quantiles
```

```python
df[['age', 'age_group']].sample(10)
```

# Create a column `purchase_frequency_days`

```python
frequency_map = {'Fortnightly': 14, 'Weekly': 7, 'Monthly': 30, 'Bi-Weekly': 15,
                 'Quarterly': 91, 'Annually': 365, 'Every 3 Months': 90, 'Daily': 1}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_map) # Map frequency to days
```

```python
df[['frequency_of_purchases', 'purchase_frequency_days']].sample(10)
```

```python
df['purchase_frequency_days'].isna().sum()  # Check for missing values in the new column
```

# Check for redundant columns and drop them

```python
(df['promo_code_used'] == df['discount_applied']).all()  # Verify consistency between promo code usage and discount applied
```
